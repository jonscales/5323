```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
price = pd.read_pickle("yahoo_price.pkl")
print(price)
```

```python
df = pd.DataFrame(price)
print(df)
```

```python

df.plot(y=["IBM","MSFT"])

plt.xlabel(' Year (daily data points) ')
plt.ylabel(' Stock Price ($) ')
plt.title('Stock Prices for IBM & MSFT')
plt.grid()

plt.show()
```
