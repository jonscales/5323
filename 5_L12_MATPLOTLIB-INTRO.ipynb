## MATPLOTLIB
Creating our data

```python
import matplotlib.pyplot as plt
import numpy as np

# this option in the code ensures that all our plots
# are displayed as we run each cell
%matplotlib inline 

x = np.linspace(0,10)  # creating samples from 0 to 10  
y = np.cos(x)
```

```python
plt.plot(x,y)
```

## MULTIPLE PLOTS

```python
x = np.linspace(0,10)  # creating samples from 0 to 10  - X-AXIS
y = np.cos(x)          # VALUES FOR 1ST CURVE
z = np.sin(x)          # VALUES FOR 2ND CURVE

plt.plot(x,y,x,z)   # x(x-axis), y(values 1st curve), z(values 2nd curve)
plt.show()
```

## LABELING AXIS & ADDING A TITLE

```python
x = np.linspace(0,10)  # creating samples from 0 to 10  - X-AXIS
y = np.cos(x)          # VALUES FOR 1ST CURVE
z = np.sin(x)          # VALUES FOR 2ND CURVE

plt.plot(x,y,x,z)   # x(x-axis), y(values 1st curve), z(values 2nd curve)
plt.xlabel(' Time ')
plt.ylabel(' Magnitude ')
plt.title('Sine and Cosine waves')

plt.show()
```

## ADDING A LEGEND

The objective is to add information that will allows us to distinguish one curve from the other

```python
x = np.linspace(0,10)  # creating samples from 0 to 10  - X-AXIS
y = np.cos(x)          # VALUES FOR 1ST CURVE
z = np.sin(x)          # VALUES FOR 2ND CURVE

plt.plot(x,y,x,z)   # x(x-axis), y(values 1st curve), z(values 2nd curve)
plt.xlabel(' Time ')
plt.ylabel(' Magnitude ')
plt.title('Sine and Cosine waves')
plt.legend(['y', 'z'], loc = 0)  # play with loc and see what happens. 
                                 # loc = 0 means best location
plt.show()
```

## SUBPLOTS

```python
x = np.linspace(0,10)  # creating samples from 0 to 10  - X-AXIS
y = np.cos(x)          # VALUES FOR 1ST CURVE
z = np.sin(x)          # VALUES FOR 2ND CURVE

#============================================
plt.subplot(1,2,1)            # A 1x2 matrix, place this one in position 1
plt.plot(x,y)   
plt.xlabel(' Time ')
plt.ylabel(' Magnitude ')
plt.title('Sine wave')
plt.legend(['sin(x)'], loc = 1)  # play with loc and see what happens. 
                                 
#============================================
plt.subplot(1,2,2)             # A 1x2 matrix, place this one in position 1
plt.plot(x,z)   
plt.xlabel(' Time ')
plt.ylabel(' Magnitude ')
plt.title('Cosine wave')
plt.legend(['cos(x)'], loc = 1)  # play with loc and see what happens. 
#============================================
plt.tight_layout()      # this option is to avoid overlapping plots
                        # you are strongly advised to use it when working
                        # with subplots.
                        # This option will ensure that all subplots are well spaced.
                        # Comment this line out, run the cell and observe the changes
plt.show()
```

## COLOR AND MARKERS FOR CURVES
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = np.linspace(0,10)  # creating samples from 0 to 10  - X-AXIS
y = np.cos(x)          # VALUES FOR 1ST CURVE
z = np.sin(x)          # VALUES FOR 2ND CURVE

#============================================
plt.subplot(1,2,1)            # A 1x2 matrix, place this one in position 1
plt.plot(x,y, 'r-x')          # red color for the line - x marker for each sample 
plt.xlabel(' Time ')
plt.ylabel(' Magnitude ')
plt.title('Sine wave')
plt.legend(['sin(x)'], loc = 1)  # play with loc and see what happens. 
                                 
#============================================
plt.subplot(1,2,2)             # A 1x2 matrix, place this one in position 1
plt.plot(x,z, 'g-o')           # green color for the line - o marker for each sample 
plt.xlabel(' Time ')
plt.ylabel(' Magnitude ')
plt.title('Cosine wave')
plt.legend(['cos(x)'], loc = 1)  # play with loc and see what happens. 
#============================================
plt.tight_layout()      # this option is to avoid overlapping plots
                        # you are strongly advised to use it when working
                        # with subplots.
                        # This option will ensure that all subplots are well spaced.
                        # Comment this line out, run the cell and observe the changes
plt.show()
```
