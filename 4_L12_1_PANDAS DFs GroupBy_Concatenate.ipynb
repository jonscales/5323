## CONCATENATE
A) It is a method used to combine multiple data frames or data sets into a single one  <br>
B) Allows data sets to be combined into a new data set. This can be done row or column-wise.
HOWEVER, this condition MUST be fulfilled --> The DIMENSIONS of your data frames MUST be the same along the axis of concatenation.  <br>
For example:<br>
B.1) ROW-WISE concatenation requires the two data frames to have THE SAME number of COLUMNS. I WANT TO HAVE MORE ROWS <br>
B.2) COLUMN-WISE concatenation requires the two data frames to have THE SAME # or ROWS. I WANT TO HAVE MORE COLUMNS

```python
d1 = {'A': 'A1 A2 A3'.split(),
      'B': 'B1 B2 B3'.split(),
      'C': 'C1 C2 C3'.split()}

d2 = {'A': 'A4 A5 A6'.split(),
      'B': 'B4 B5 B6'.split(),
      'C': 'C4 C5 C6'.split()}

df1 = pd.DataFrame(d1, index='1 2 3'.split())
df2 = pd.DataFrame(d2, index='4 5 6'.split())

#let's concatenate them
# ROW - WISE CONCATENATION
df3 = pd.concat([df1, df2])  
df3
```

```python
# COLUMN - WISE CONCATENATION
df4 = pd.concat([df1, df2], axis=1, sort=True)  
df4
```

## Let's introduce a minor modification

```python
df2.index = '1 2 3'.split()  # here we are changing the index of df2 
                             # to match the index of df1
df5 = pd.concat([df1, df2], axis=1)  
df5
```

## let's explore one more tiny change - rename the second group of A B Cs

```python
df2.columns = 'D E F'.split()
df6 = pd.concat([df1, df2], axis=1)  
df6
```
